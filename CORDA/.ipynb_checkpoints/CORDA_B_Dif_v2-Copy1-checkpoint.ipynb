```python
# For the normally distributed genes, we will perform a hierarchical clustering to decide which genes could potentially be
# labeled as expressed when creating our reconstruction with CORDA.

# For each cluster, we will calculate which gene (Gcv) has the biggest coefficient of variation (Gcv_mean/Gcv_std). 
# Based on this coefficient, we will binarize expression data of the genes in the cluster. 
# (i.e. if 'value' > mean(Gcv_mean) for each 'value' of each gene then we label that value as 1, if not 0)
# For each gene, we will get a vector (binGene) with 0s and 1s instead of the exp. levels. We will then calculate mean(binGene) 


# Directory
Data=[] 
names = []
for i in X2.keys():
    names.append(i)
    Data.append(np.float32(X2[i]))
```

```python
# We build a dendogram to select how many clusters to use for the classification

dendrogram = sch.dendrogram(sch.linkage(Data, method = 'ward'))
#val = np.array(list(genes_median.items()))[:,1]
#val = val.astype(float)
# dendrogram = sch.dendrogram(sch.linkage(val.astype(float), method = 'ward'))
```

```python
# We use Agglomerative Clustering with 3 clusters
# each one of the cluster sets the low, medium and high confidence respectively.

from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(Data)
# y_hc = hc.fit_predict(val)
```

```python
print (np.shape(names),np.size(y_hc))
```

```python
## Creating a data frame 

Data_Info = {'Genes':np.array(names),'Cluster':np.array(y_hc+1)}
Data_Info = pd.DataFrame(Data_Info)
Data_Info[1:30]
```

```python
# We organize the data from the clusters so we can plot it and observe it

zero=[]
uno=[]
dos=[]
#tres=[]
#cuatro=[]
for x,i in enumerate(y_hc):
    if i==0:
        zero.append((mode(Data[x])[0][0], np.mean(Data[x]),np.std(Data[x])))
    if i==1:
        uno.append((mode(Data[x])[0][0], np.mean(Data[x]),np.std(Data[x])))
    if i==2:
        dos.append((mode(Data[x])[0][0], np.mean(Data[x]),np.std(Data[x])))
#    if i==3:
#        tres.append((np.mean(Data[x]),np.std(Data[x])))
#    if i==4:
#        cuatro.append((np.mean(Data[x]),np.std(Data[x])))
```
